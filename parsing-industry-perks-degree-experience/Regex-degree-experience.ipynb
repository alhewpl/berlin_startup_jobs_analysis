### Transform csv into json for more easily manipulation

```python
data = []
with open('../Machine-Learning-clustering-skills/DB/merged_jobs.csv', 'r', encoding="utf-8") as file:
    for row in csv.DictReader(file):
            data.append(row)

    json_data = json.dumps(data)
    output_file=open('jobs.json', 'w')
    output_file.write(json_data)
    output_file.close()

    
with open ('jobs.json', 'r') as in_file:
    lines = in_file.readlines()
    for line in lines:
        content = json.loads(line) 
        #print(content)    
                    
```

### Connect to db and add values to separate table

```python
conn = sqlite3.connect('/Users/alina/Desktop/StartupJobs/jobs')
```

```python
description = {d['ID']: d['description'] for d in data2}
for k, v in description.items():
    v = v.lower()
    v = unicodedata.normalize("NFKC", v) #remove extended ascii
    
    if  any((i in v for i in degree )) or re.search(education, v, flags=re.U):
        req = 'yes'
    else:
        req = 'no'
    c = conn.cursor()
    c.execute("INSERT INTO degree(job_id, required, type) VALUES (?,?,?)",(int(k), req, '...'))
    
    if any((p in v for p in exp)) or re.search(experience, v, flags=re.U) or re.search(exper, v, flags=re.U):
        e = 'yes'
    else:
        e = 'no'
    #c = conn.cursor()
    c.execute("INSERT INTO required_experience(job_id, require_experience) VALUES (?,?)", (int(k), e))
```

```python
conn.commit()
conn.close()
```
